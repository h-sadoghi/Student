## Research and Technology Background and Introduction to Specialized Lab
Dr. Hadi Sadoghi Yazdi's research background is in the areas of pattern recognition, machine learning, computer vision, signal processing. He established the specialized Pattern Recognition Lab with the aim of conducting research and technological activities. He is the director and manager of the lab.

## Educational Background
Dr. Hadi Sadoghi Yazdi graduated with a bachelor's degree in Electrical Engineering - Electronics from Ferdowsi University of Mashhad, and received his master's and Ph.D. degrees from Tarbiat Modares University. He has been a faculty member in the Department of Computer Engineering at the Faculty of Engineering, Ferdowsi University, since 2007.
